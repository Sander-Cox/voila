## Introduction
COVID-19 has been ravaging all parts of the world for over 2 years now. Many lives have been changed in permanent ways and almost every country has had a different strategy of dealing with the pandemic. It could be massively beneficial to look back at the results of these policies of different western countries to see what worked and what did not. 
Research question: How do different western countries fare with regards to weekly COVID hospital admissions per million? 

I hope to anwer this question by comparing multiple western countries in regards to the weekly hospital admissions per million. This way all countries would be easier to compare despite of large differences in size. 


## Method
In order to anwer this research question the entire up-to-date COVID-19 dataset was loaded. Multiple columns were extracted that could be of interest in this, or further research including: 
"iso_code", "location", "date", "new_deaths_per_million", "weekly_icu_admissions_per_million", "weekly_hosp_admissions_per_million", "people_fully_vaccinated_per_hundred" 

Then, using descriptive statistics ( .describe(), .value_counts() ) it was determined that the variable "new_deaths_per_million" included a few negative admissions which is impossible. Hence these were corrected for by ensuring all variables are positive by making them absolute. There were no further outliers.



## Results
The included countries are in alfabetical order are: Belgium, France, Germany, Netherlands, United Kingdom, United States.

```python
# First I import the relevant modules
import pandas as pd
import altair as alt 
import ipywidgets as widgets
from ipywidgets import interact
from ipywidgets import fixed


# Then I imported the COVID-19 dataset and saved the dataframe as 'df'
df = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv", parse_dates=['date'])


# I created two variables in order to extract the relevant columns from the dataset and saved it in the same variable
included_countries = ['NLD', 'BEL', 'GBR', 'FRA', 'USA', 'DEU']
included_columns = ["iso_code", "location", "date", "new_deaths_per_million", 
             "weekly_icu_admissions_per_million", "weekly_hosp_admissions_per_million", 
             "people_fully_vaccinated_per_hundred"]
df = df.loc[df.iso_code.isin(included_countries), included_columns]
df = df.dropna(thresh=3, subset=["new_deaths_per_million", "weekly_icu_admissions_per_million", "weekly_hosp_admissions_per_million", "people_fully_vaccinated_per_hundred"])

# Lastly I cleaned the few negative values in 'new_deaths_per_million' by making all variables absoluut (positive)
df['new_deaths_per_million'] = abs(df['new_deaths_per_million'])


#df.describe()
#df.value_counts()

df.agg({
    'new_deaths_per_million': ['mean', 'median', 'std'],
    'weekly_icu_admissions_per_million': ['mean', 'median', 'std'],
    'weekly_hosp_admissions_per_million': ['mean', 'median', 'std']
})
```

Then a pair of interactive graphs was created where the user can select a period of time in the line graph and the bar graph will change or the user can select one or, using shift, multiple countries from the bar graph to visualise them in the line-graph.

```python
# Two variables that will be later referred back to in order to be able to select a partition of the dates
multi = alt.selection_multi(fields=['location'])
interval = alt.selection_interval(encodings=['x'], empty='all')

# The creation of the line-graph
line = alt.Chart(df).mark_line().encode(
    x=alt.X('date:T'),
    y='weekly_hosp_admissions_per_million',
    color=alt.condition(multi & interval, 'location', alt.value('lightgray'))
). properties(
    selection=interval,
    title='Weekly hospital admissions per milion over time of different western countries'
)

# The creation of the bar-graph
bar = alt.Chart(df).mark_bar().encode(
    x='average(weekly_hosp_admissions_per_million)',
    y='location',
    color=alt.condition(multi, 'location', alt.value('lightgray'))
).properties(
    selection=multi,
).transform_filter(
    interval
)

# The displaying of both bars
line & bar
```

##Conclusion
Looking at the graphs it becomes rather clear that countries like Germany and the Netherland are able to keep weekly hospital admissions considerably lower than the United States. This means the US could look at some of the policies these countries have implomented in order to better fight the pandemic in the future.
